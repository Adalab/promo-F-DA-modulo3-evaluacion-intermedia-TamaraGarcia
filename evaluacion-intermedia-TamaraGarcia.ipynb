#### PREPARACIÓN DE LOS DATOS
---

```python
# Cambiar los nombres de las columnas a minúscula
nuevas_columnas = {columna: columna.lower() for columna in df.columns}
df.rename(columns = nuevas_columnas, inplace = True)
df
```

```python
# Función para cambiar comas por puntos

def funcion(elemto):
    try:
        return float(elemto.replace(",", "."))
    except:
        return np.nan


columnas = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']
for columna in columnas:
    df[columna] = df[columna].apply(funcion)

df

# Para comprobar que se hayan cambiado, checkeo el df.dtypes
```

```python
# Cambiar las mayúsculas a minúsculas en las columnas

columnas = ['title', 'album']
for elemto in columnas:
    df[elemto] =df[elemto].str.lower()

df
```

```python
# Para localizar cuales son las que tienen $
df.loc[df['artist'].astype(str).str.contains('\$')]

for elemto in df['artist']:
    df[elemto] = df[elemto].str.replace('$','')
```

Ahora queremos analizar la colaboración entre artistas en la industria musical, se tienen datos de las colaboraciones realizadas en los últimos años. Se desea conocer cuáles son los artistas que han colaborado en mayor cantidad de canciones y generar un ranking con los diez primeros.

¿Qué artistas con los que más colaboraciones tienen? 

```python
# filtro
colabs = df[df['track'].str.contains('feat')]

# cuántas canciones x artista
canciones = colabs.groupby('artist')['track'].count().sort_values(ascending= False).reset_index()
canciones
```

```python
# Los 10 que más han colaborado

canciones.iloc[1:11,:]
```

```python
# canciones con el artiste del primer lugar
canciones.iloc[1,:]
```

```python
# canciones con el artiste del décimo lugar
canciones.iloc[10,:]
```
